<p>Determine which games would have been possible if the bag had been loaded with only 12 red cubes, 13 green cubes, and 14 blue cubes. <em>What is the sum of the IDs of those games?</em></p>

```python
import pandas as pd
import re
from typing import List
```

```python
def possibleOrNot(arr:List[str], r,g,b):
    for i in arr:
        spl = i.strip().split(' ') if i != '' else ['','']
        if spl.count('green'):
            if int(spl[0]) > g:
                return False
        elif spl.count('blue'):
            if int(spl[0]) > b:
                return False
        elif spl.count('red'):
            if int(spl[0]) > r:
                return False
    return True
```

```python
df = pd.read_table(sep=';', filepath_or_buffer='input.txt', names=range(6), )
df['Game'] = df[0].str.split(':').apply(lambda x: x[0]) # gotta be a better way to do that
df[0] = df[0].str.split(':').apply(lambda x: x[1:][0]) # gotta be a better way to do that
df=df.fillna('')
```

```python
cols = df.columns[:-1]
copyDF = df.copy(deep=True)
for i in cols:
    copyDF[i] = copyDF[i].apply(lambda x: x.split(','))
    # copyDF[i] = copyDF[i].apply(lambda x: [y for y in x.split(' ')])
copyDF
    
```

```python
slice = copyDF.iloc[:,:-1].apply(lambda s: s.apply(lambda x: possibleOrNot(x,r=12,g=13,b=14)))
slice
```

```python
checkForTrue = slice.all(axis=1).reset_index()
checkForTrue['index'] = checkForTrue['index']+1
checkForTrue
```

```python
allTrue = checkForTrue.loc[checkForTrue[0]==True]
allTrue['index'].sum()
```
